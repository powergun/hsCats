# Function Types

_P/134 what is a function type_

a function type $a \rightarrow b$ is a set of 
morphisms between objects $a$ and $b$

a set of morphisms between two objects in any category
is called a **hom-set**, which is an object in the Set category.

_P/136 universal construction_

a function type may be considered a composite type because
 of its relationship to the argument type and the result type.

that is the pattern: **a product of two objects $z$ and $a$
connected to another object $b$ by a morphism $g$**

_P/144 exponentials_

in mathematical literature, the function object, or the
internal hom-object between two objects $a, b$, is often
called the **exponential** and denoted by $b^{a}$

**notice that the argument type is in the exponent**

for instance a (pure) function from $Bool$ is completely
specified by a pair of values: one corresponding to $False$
and one corresponding to $True$

the set of all possible functions $Bool \rightarrow Int$ is
the set of all pairs of Ints

this is the same as the product $Int \times Int$ or, $Int^{2}$

_P/146 cartesian closed categories_


a cartesian closed category must contain

- the terminal object
- a product of any pair of objects
- an exponential for any pair of objects

cartesian closed categories provide models for the simply
typed lambda calculus, which forms the basis of all typed
programming languages
